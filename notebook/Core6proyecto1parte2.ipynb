Filtrado de datos

```python
print("5. FILTRADO DE DATOS")
print("-" * 30)

print("Filtro 1: Ventas > 50 ")
na_sales_filter = df[df['Total Amount'] > 50]
print(f"✓ Se encontraron {len(na_sales_filter)} Ventas")
print("-" * 50)
na_sales_filter.head(10)
```

```python
print("Filtro 2: Precio < 50:") #en el data sen no existen valores menores a 0.5 se ocupara menor a 50 en su defecto
jp_sales_filter = df[df['Price per Unit'] < 50]
print(f"✓ Se encontraron {len(jp_sales_filter)} Productos")
print("-" * 50)
```

```python
print("Filtro 3: Electronics con ventas  > 2 (usando query()):")
Product_Category_filter = df.query("`Product Category` == 'Electronics' and `Quantity` > 2")
print(f"✓ Se encontraron {len(Product_Category_filter)} Productos")
print("-" * 50)
Product_Category_filter.head(10)
```

```python
print("6. SLICING DE DATOS")
print("-" * 30)

# Selección 1: Selecciona y muestra solo las columnas producto y ventas del DataFrame
print("Selección 1: Solo las columnas producto  y ventas:")
print("-" * 40)
name_global_sales = df[['Product Category', 'Total Amount']]
name_global_sales.head(10)
```

```python
print("Selección 2: Usando loc[] - filas 5 a 10 (inclusive), columnas Producto y ventas:")
print("Nota: loc[] usa índices basados en 0, pero incluye el final")
print("-" * 60)
loc_selection = df.loc[4:9, ['Product Category', 'Total Amount']]  # Índices 4 a 9 = filas 5 a 10
loc_selection
```

```python
print("Selección 3: Usando iloc[] - primeras 5 filas y primeras 3 columnas:")
print("-" * 60)
iloc_selection = df.iloc[0:5, 0:3]  # Filas 0-4 (5 filas), columnas 0-2 (3 columnas)
iloc_selection
```
